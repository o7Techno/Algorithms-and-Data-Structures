# Анализ производительности алгоритмов сортировки строк

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set(style="whitegrid")
```

## Загрузка данных

```python
df = pd.read_csv("results.csv")

df.head()
```

## Время выполнения алгоритмов (микросекунды)

```python
cases = df['Case'].unique()

for case in cases:
    sub = df[df['Case'] == case]
    plt.figure(figsize=(8, 4))
    sns.lineplot(data=sub, x="N", y="AvgTime_us", hue="Algorithm", marker="o")
    plt.title(f"Время сортировки — {case}")
    plt.xlabel("Размер массива (N)")
    plt.ylabel("Время (μs)")
    plt.legend(title="Алгоритм")
    plt.tight_layout()
    plt.show()
```

## Посимвольные сравнения

```python
for case in cases:
    sub = df[df['Case'] == case]
    plt.figure(figsize=(8, 4))
    sns.lineplot(data=sub, x="N", y="AvgComps", hue="Algorithm", marker="o")
    plt.title(f"Сравнения символов — {case}")
    plt.xlabel("Размер массива (N)")
    plt.ylabel("Количество сравнений")
    plt.legend(title="Алгоритм")
    plt.tight_layout()
    plt.show()
```
